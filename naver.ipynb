```python
import pandas as pd

# 1. 데이터프레임 생성
data = {"금액": ["5억 8,980", "6억 1,488", "6억 1,637", "6억 1,638", 
                 "6억 120", "6억 2,152", "6억 2,220", "6억 2,339", "6억 2,545"]}
df = pd.DataFrame(data)

# 2. 문자열을 숫자로 변환하는 함수
def convert_to_number(value):
    # "억"과 "만원" 단위 분리
    value = value.replace("억", "").replace(",", "")  # "5억 8,980" -> "5980"
    parts = value.split()  # 공백 기준으로 분리
    # 억 단위 처리
    if len(parts) == 2:
        first = int(parts[0]) * 10000  # 억 단위 * 10,000
        second = int(parts[1])  # 만원 단위
        return first + second  # 총 금액
    else:
        return int(parts[0]) * 10000  # 만약 "6억"만 있을 경우

# 3. 금액을 숫자로 변환하여 새로운 컬럼에 추가
df["금액(만원)"] = df["금액"].apply(convert_to_number)

# 4. 금액을 기준으로 정렬
df_sorted = df.sort_values(by="금액(만원)", ascending=True).reset_index(drop=True)

# 5. 결과 출력
print(df_sorted)
```
